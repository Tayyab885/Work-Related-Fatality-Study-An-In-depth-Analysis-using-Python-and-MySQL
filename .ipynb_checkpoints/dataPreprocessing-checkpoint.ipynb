#### EDA:

```python
plan = df['plan'].value_counts()
plan
```

```python
index = plan.index
value = plan.values
plt.figure(figsize=(14,7))
ax = sns.barplot(x = index, y = value, palette='rainbow')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Safty Plans')
plt.xlabel('Plan')
plt.ylabel('Count')
plt.show()
```

```python
citation = df['citation'].value_counts()
citation
```

```python
index = citation.index
value = citation.values
plt.figure(figsize=(14,7))
ax = sns.barplot(x = index, y = value, palette='rainbow')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Citations Issued')
plt.xlabel('Citation')
plt.ylabel('Count')
plt.show()
```

```python
state = df['state'].value_counts()
state
```

```python
index = state.index
value = state.values
plt.figure(figsize=(12,16))
ax = sns.barplot(x = value, y = index, palette='rainbow')
for i in ax.containers:
    ax.bar_label(i)
plt.title('No. of States')
plt.ylabel('States')
plt.xlabel('Count')
plt.show()
```
